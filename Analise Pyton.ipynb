Renomeando Colunas

```python
dados1.columns = ['Loja','Data','Temperatura','Preço_Combustivel','Desconto1','Desconto2','Desconto3','Desconto4','Desconto5','indice_preço','Desemprego','Feriado']
dados1.head()
```

```python
dados2.columns = ['Loja','Departamento','Data','Vendas_Semanais','Feriado']
dados2.head()
```

```python
dados3.columns = ['Loja','Tipo','Tamanho']
dados3.head()
```

```python
dados1['Data'] = pd.to_datetime(dados1['Data'], dayfirst=True)
dados1.dtypes
```

```python
dados2['Data'] = pd.to_datetime(dados2['Data'], dayfirst=True)
dados2.dtypes
```

```python
dados3.dtypes
```

```python
dados1.fillna(0, inplace=True)
```

```python
tabela = pd.merge(dados2, dados3, how='inner', on='Loja')
tabela.head()
```

```python
tabela = pd.merge(tabela, dados1, how='inner', on=['Loja', 'Data'])
```

```python
num_linhas_total = tabela.shape[0]
print(f"O DataFrame tem {num_linhas_total} linhas.")
```

```python
tabela.nunique()
```

```python
tabela.head()
```

```python
media = sts.mean(tabela['Vendas_Semanais'])
mediana = sts.median(tabela['Vendas_Semanais'])
minimo = min(tabela['Vendas_Semanais'])
max = max(tabela['Vendas_Semanais'])
print(media)
print(mediana)
print(minimo)
print(max)
```

```python
loja = tabela.groupby(['Loja'])['Vendas_Semanais'].sum()
loja.sort_values(ascending=False)
```

```python
loja = tabela.groupby(['Tipo'])['Vendas_Semanais'].sum()
loja.sort_values(ascending=False)
```

```python
loja = tabela.groupby(['Departamento'])['Vendas_Semanais'].sum()
loja.sort_values(ascending=False)
```

```python
# Criando novas colunas a partir da coluna 'Data'
tabela['Ano'] = tabela['Data'].dt.year
tabela['Mês'] = tabela['Data'].dt.month
tabela['Nome_Mês'] = tabela['Data'].dt.month_name()
tabela['Dia_Semana'] = tabela['Data'].dt.day_name()
```

```python
srn.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
vendas_mensais = tabela.groupby(['Ano', 'Mês'])['Vendas_Semanais'].sum().reset_index()
srn.lineplot(data=vendas_mensais, x='Mês', y='Vendas_Semanais', hue='Ano', marker='o')
plt.title('Vendas Mensais por Ano')
plt.xticks(range(1,13))
plt.xlabel('Mês')
plt.ylabel('Total de Vendas Semanais')
plt.show()
```

```python
dias_ordem = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
vendas_dia = tabela.groupby('Dia_Semana')['Vendas_Semanais'].mean().reindex(dias_ordem)
vendas_dia.plot(kind='bar', color='skyblue')
plt.title('Média de Vendas por Dia da Semana')
plt.ylabel('Vendas Semanais Médias')
plt.xlabel('Dia da Semana')
plt.xticks(rotation=45)
plt.show()
```

```python
vendas_tempo = tabela.groupby('Data')['Vendas_Semanais'].sum()
vendas_tempo.plot()
plt.title('Tendência Geral das Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Total de Vendas Semanais')
plt.show()
```
